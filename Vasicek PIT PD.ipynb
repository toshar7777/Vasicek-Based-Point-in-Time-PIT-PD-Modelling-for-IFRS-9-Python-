For Fitted PIT Score:-

```python
rho=0.01 #seed value
```

```python
threshold= norm.ppf(TTC_PD)
threshold
```

```python
# Z range seed value:-
np.random.seed(42)
df['z']=np.random.normal(0,1,len(df))    #mean--> 0 , std dev--> 1 and no of values--> len(df)
```

```python
# Fitted PIT SCORE:-
df['fitted_PIT_Score']=(threshold-np.sqrt(rho)*df['z'])/np.sqrt(1-rho)
df
```

```python
df['error']=(df['actual_PIT_score']-df['fitted_PIT_Score'])**2
```

# Optimization using least square

1) Here we have to optimize Z values and rho.
2) For optimization using least squares we have to give the synatax as: least_square(function,initial parameters, bounds)

```python
# Initial Parameters (give it as a list/array):-
initial_params=np.concatenate(([rho],df['z'].values))
initial_params
```

```python
def objective(params):
    rho= params[0]
    z=params[1:]
    fitted_PIT_Score =(threshold-np.sqrt(rho)*z)/np.sqrt(1-rho)
    residuals= df['actual_PIT_score'] - fitted_PIT_Score
    return residuals
```

Constraints:-

1) z value range --> (-7 to +7)
2) rho >=0 (regulatory requirement)
3) variance of z = 1 (in least squares, it automatically makes the variance constant or 1.
4) The above constraints(bounds) can be created seperately or given while initializing, here we are giving seperately.

```python
# Bounds LB for both z and rho--> (0,-7) UB for z and rho--> (1,+7)
bounds=([0]+[-7]*len(df),[1]+[7]*len(df))
bounds
```

```python
from scipy.optimize import least_squares
```

```python
results= least_squares(objective,initial_params,bounds=bounds)
```

```python
optimal_rho=results.x[0]
optimal_rho
```

```python
# Optimal Z values in place of old z values:-
df['z']=results.x[1:]
df
```

```python
# New Fitted_PIT_Score in place of Old Fitted_PIT_Score:-
df['fitted_PIT_Score']=(threshold-np.sqrt(optimal_rho)*df['z'])/np.sqrt(1-optimal_rho)
df
```

```python
# New error:-
df['error']=(df['actual_PIT_score']- df['fitted_PIT_Score'])**2
df
```

```python
final_SSE=df['error'].sum()
final_SSE
```

# Regress z with the lagged values of macro eco variables

```python
MEV_df=pd.DataFrame()
```

```python
MEV_df['z']=df['z']
```

```python
MEV_df['Lag_HPI']=df['HPI'].shift(1)
MEV_df['Lag_UR']=df['UR'].shift(1)
MEV_df
```

```python
# drop the 1st row, as it has null values:-
MEV_df.drop(0,axis=0,inplace=True)
```

```python
MEV_df
```

```python
# Regress Z with Lagged values of HPI and UR:-
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
from statsmodels.stats.anova import anova_lm
```

```python
model = ols('z ~ Lag_HPI + Lag_UR',data=MEV_df).fit()
```

```python
model.summary()
```

```python
anova_results=anova_lm(model)
anova_results
```

```python
# We have the forecast of HPI and UR give by the macro eco analysis team:-

forecast_base=pd.read_excel('forecast.xlsx',sheet_name='Sheet1')
forecast_base
```

```python
forecast_base.rename(columns={'HPI' :'Lag_HPI', 'UR' : 'Lag_UR'},inplace=True)
```

```python
forecast_base
```

```python
# Predict the Z values:-
MEF_pred=model.predict(forecast_base)
MEF_pred
```

```python
var_of_residual=anova_results.loc['Residual','mean_sq']
var_of_residual
```

```python
# Find the Forecasted PIT FL PD using vasicek:-
forecast_base['vasicek']=norm.cdf((threshold - np.sqrt(optimal_rho)*MEF_pred)/(np.sqrt(1-optimal_rho + optimal_rho * var_of_residual)))
```

```python
forecast_base
```

Forecasted PIT FL PD for the next quarter is 0.046568.

# Now we calibrate using the log odds shift calibration:-


We have PIT FL PD for each snapshot, but we want PIT FL PD for each rating grade. TO adjest the PD curve based on target central tendency.

```python
df1=pd.read_excel('1_TTC_PD.xlsx',sheet_name='TTC_PD')
df1
```

```python
TTC_PD=df1.groupby('Grade')[['Default','Pd_pred']].mean().reset_index()
TTC_PD
```

```python
TTC_PD.rename(columns={'Default': 'ODR','Pd_pred': 'Model_PD'},inplace=True)
TTC_PD
```

```python
TTC_PD['ODR']=TTC_PD['ODR'].apply(lambda x: 0.0003 if x==0 else x)
TTC_PD
```

```python
distn= df1.groupby('Grade').size().reset_index(name='count')
distn
```

```python
TTC_PD['%_Population']=distn['count']/distn['count'].sum()
TTC_PD
```

```python
TTC_pd=(TTC_PD['ODR']*TTC_PD['%_Population']).sum()
TTC_pd
```

```python
temp=TTC_PD[['Grade','ODR','%_Population']]
temp
```

```python
temp['Log_Odds']=np.log(temp['ODR']/(1-temp['ODR']))
temp
```

```python
from scipy.stats import linregress
```

```python
slope,intercept,r,pvalue,std_error=linregress(x=temp['Grade'],y=temp['Log_Odds'])
```

```python
slope
```

```python
intercept
```

```python
# intercept seed value:-
seed_intercept=-8
```

```python
temp['fitted_log_odds']=seed_intercept + slope*temp['Grade']
temp
```

```python
temp['Annual_Cal_PD']=np.exp(temp['fitted_log_odds'])/(1+np.exp(temp['fitted_log_odds']))
temp
```

```python
TTC_PD_temp=(temp['Annual_Cal_PD']*temp['%_Population']).sum()
TTC_PD_temp
```

```python
tgt_pd=forecast_base['vasicek'][0]
tgt_pd
```

```python
from scipy.optimize import minimize_scalar
```

```python
def obj_functn(adj_intercept):
    temp['fitted_log_odds']=adj_intercept + slope*temp['Grade']
    temp['Annual_Cal_PD']=np.exp(temp['fitted_log_odds'])/(1+np.exp(temp['fitted_log_odds']))
    TTC_PD_temp=(temp['Annual_Cal_PD']*temp['%_Population']).sum()
    return abs(TTC_PD_temp - tgt_pd)
```

```python
result1=minimize_scalar(obj_functn)
```
